```
#데이터 불러오기

import pandas as pd
fifa2020 = pd.read_csv('') #경로 채워놓기
df = pd.DataFrame.copy(fifa2020.sort_values(by='overall', ascending=False).head()) #종합능력을 기준, 높은 순서대로 1000명을 불러온다.

test_features = ['long_name', 'physic', 'dribbling', 'passing', 'shooting']
test_df = pd.DataFrame(df, columns = test_features) #데이터 프레임 (데이터를 분석하기 위한 형태로 행과 열(속성)으로 이루어진다.)
print(test_df.head(5))

# Jan Oblak은 왜 값이 NAN일까?
# 군집을 만들 때 비어 있는 값이 있으면 좋은 결과가 나올까?
```

```
test_df = test_df.dropna(axis=0)

# 불러온 데이터에서 값이 없는(NaN)을 삭제해준다
```

```
print(test_df.head(5))
```

```
#학습 데이터 준비하기
# numpy 행렬 연산, 벡터 연산에 필요한 도구
import numpy as np

XY = np.array(test_df) #행렬로 test_df의 값을 가져온다
X = XY[:,:] #X값 우리가 군집을 만들 값으로 1번인 physic, 3미만이므로 2번인 dribblong을 가져온다
```

```
# K-평균 군집화 알고리즘 (K-Mean 알고리즘 적용하기)
# K를 3으로 하고 여러번 실행해보자
# 왜 체력하고 드리블만? X,Y 좌표 평면에 나타내기 위해서

k = 3
C_x = np.random.choice(X[:,0], k) #체력에서 랜덤으로 3개 가져옴
C_y = np.random.choice(X[:,1], k) #드리블에서 랜덤으로 3개 가져옴 
C = np.array(list(zip(C_x,C_y))) #체력값과 드리블값을 하나의 행렬로 묶어줌
print(C)

#여러번 시작해보자 
```

```
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm


Physic = test_df['physic']
Dribbling= test_df['dribbling']
plt.figure(figsize=(10,5))
plt.title('physic%dribbling')
plt.xlabel('physic')
plt.ylabel('Dribbling')


plt.scatter(,,marker = '^', c = 'blue', s = 10, label = 'players') #선수의 분포 X축을 체력으로, Y축을 드리블 능력이 되도록 채워보자
plt.scatter(C_x, C_y, marker = '*', s = 200, c = 'black',label = 'centroids') #무작위 중심 좌표 산점도 
plt.legend(loc = 'best') #범례의 위치를 알아서 최적으로 맞춰줌
plt.grid( )
plt.show( )
```

```
#두 점 사이의 거리 (유클리디안 거리 계산 함수 만들기)

def Distance(A, B):
  return np.sqrt(np.sum(np.power((A-B),2))) #sqrt : 루트, power : 제곱
```

image.png

```
C_old = np.zeros(C.shape) #군집의 중심과 동일한 크기의 행렬 선언
clusters = np.zeros(len(X)) #선수 데이터만큼의 행렬선언 중심과 선수의 위치간의 거리 계산
flag = Distance(C,C_old) #현재 클러스터(C)의 위치와 C_old(초기값 0,0)사이의 거리를 나타내는 flag
print(C_old)
print(flag)
```

```
from copy import deepcopy #완벽한 복사를 위해 불러오는 도구 

distances = [] #각각의 선수들과 클러스트간의 거리를 저장하기 위한 리스트

while flag!=0: #더 이상 클러스트를 옮길 필요가 없을 때까지 (데이터의 중앙값 자체가 클러스트일 때)
  for i in range(len(X)): #선수값의 길이만큼 처음 부터 끝까지 돌면서
    for j in range(k): #0, 1, 2번 군집과의 거리를 구함
      temp = Distance(X[i],C[j])  # 선수값과 0,1,2번 군집과의 거리
      distances.append(temp)
    cluster = np.argmin(distances) #제일 거리가 작은 군집을 찾아내
    clusters[i]=cluster #cluster[i]에 몇번 군집에 속했는지를 저장해줌 
    distances = [] #군집 결정이 됐으니 다시 빈값으로 만들어줌 

  C_old = deepcopy(C) #현재의 군집 위치를 old에 저장해놓음

  #새로운 군집 중심 구하는 부분
  for i in range(k): #0,1,2번 군집을 돌면서
    points = [X[j] for j in range(len(X)) if clusters[j]==i] #각 군집의 값을 points에 저장하고 
    C[i] = np.mean(points) #평균값을 새로운 군집위치로 만듦 

  flag = Distance(C,C_old) #현재 군집과 기존 군집과의 거리를 flag에 저장함
```

```
plt.title("FIFA 20 TOP 1000 LEVEL")
plt.scatter(X[clusters == 0, 0], X[clusters == 0, 1], s = 50, c = 'red', marker = 'o', edgecolor = 'black',label = 'High')
plt.scatter(X[clusters == 1, 0], X[clusters == 1, 1], s = 50, c = 'yellow', marker = 'x', edgecolor = 'yellow',label = 'Middle')
plt.scatter(X[clusters == 2, 0], X[clusters == 2, 1], s = 50, c = 'blue', marker = '^', edgecolor = 'blue',label = 'Low')
plt.scatter(C[:,0], C[:,1], s = 250, marker = '*', c = 'black', edgecolor = 'black', label = 'Centroids')
plt.legend( )
plt.grid( )
plt.savefig('/content/gdrive/MyDrive/PL2/FIFA 20 TOP 1000 LEVEL.png') 
plt.show( )
```
