```python
# 파일 읽어와 앞에서 18개 읽어오기.
# 왜 18일까?
df_kang = pd.read_csv("")
df_kang.head(18)
```

```python
# 고유값만 뽑아내기
# 목적? : 몇년까지의 데이터를 가지고 있나
# 강원도 시군구의 이름은?

print(df_kang['시군명'].unique())
print(len(df_kang['해당연도'].unique())) #25년동안
kang_name = []

 
```

```python
#1995년도의 강원도 시군구 인구 데이터 가져오기

year_tot = []
df_kang['인구 (명)'][0:] #항상 뒤에 오는 숫자는 의도한 숫자보다 하나 증가해야 하는 걸 잊지 말기
for i in range():
  year_tot.append(df_kang['인구 (명)'][i])

print(year_tot)
```

```python
# 1995년 인구수는 0번 부터 번까지
# 1996년 인구수는 18번 부터 번까지
# 1997년 인구수는 36번 부터 번까지 

# 2019년 인구수는 432번 부터 449까지

#시작은 18*i (i: 0부터 24까지) 
#종료는? 18*i +  -> for 루프 돌 때는 +1해주기
```

```python
year_tot = [] #1995년부터 2019년까지의 시군별 인구수를 저장할 리스트
year_one = [] #한 해의 인구수를 저장할 임시 리스트

for i in range(): 
  for i in range(i*,i*18+):
    year_one.append(df_kang['인구 (명)'][i])
  year_tot.append(year_one)
  year_one = []

print(year_tot)
```

```python
for i in range(25):
  plt.figure(figsize=(20,5))
  #plt.figure(figsize=(20,5))
  plt.title('강원도 지역별 인구수 ({})'.format(i+1995))
  plt.bar(kang_name,year_tot[i],color='blue',label='인구(명)')
  plt.legend()
  plt.show()
```

```python
from matplotlib import animation
from matplotlib import rc
#rc('animation', html='jshtml')  #GIF 파일 형식
rc('animation', html='html5') #동영상 파일 형식
from IPython.display import HTML

fig=plt.figure(figsize=(20,10))
n=24 #Number of frames
barcollection = plt.bar(kang_name,year_tot[0]) #막대 그래프 시각화의 첫번째 대표 값

def animate(i):
    plt.title("강원도 시군구 인구 변화 ({})년".format(i+1995)) #제목 
    plt.ylim(0,350000) #y값 범위
    y=year_tot[i]
    for i, b in enumerate(barcollection):
        b.set_height(y[i])

anim=animation.FuncAnimation(fig,animate,repeat=False,blit=False,frames=n,
                             interval=300)
anim
```
