```python
# 한국어 말 뭉치 (https://www.korean.go.kr/front/etcData/etcDataView.do?mn_id=46&etc_seq=71) CSV 파일 가져와 숫자 제거 후 저장


import pandas as pd
import re

#파일을 판다스 형식으로 불러와 리스트([]) 형식으로 저장

word_df = pd.read_csv('https://raw.githubusercontent.com/CarlosQuperman/sdhs202121/main/%ED%95%9C%EA%B5%AD%EC%96%B4%20%ED%95%99%EC%8A%B5%EC%9A%A9%20%EC%96%B4%ED%9C%98%20%EB%AA%A9%EB%A1%9D.csv',encoding='utf-8')
word_df.head(10)

word = word_df['단어']
word_lst = word.values.tolist()
print(word_lst)


#동음이의어를 표시하는 숫자를 제거하기 위한 코드 숫자 제거 이후 word_lst2에 저장

word_lst2 = []
no_num = re.compile('[^0-9]')

for i in word_lst[:]:
  i = "".join(no_num.findall(i))
  if i not in word_lst2:
    word_lst2.append(i)

print(word_lst2)
```
