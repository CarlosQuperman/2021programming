# TKINTER
* tkinter는 파이선에서 그래픽 환경을 만들어주고 여러 프로그래밍 과정을 연결시킬 수 있는 라이브러리(도구집)입니다.
* 여기서는 tkinter를 이용하여 섭씨 <> 화씨를 전환하는 프로그램을 만들어보고
* 이런 활동을 통해 GUI 환경에 대해 이해해보고 만들어봅시다.

# 이번 시간 활동은 코립이 아니라 파이썬 IDLE에서 합니다.

* IDLE이란? : 온라인이 아닌 오프라인(내 컴퓨터)에서 파이선 프로그래밍을 하고 프로그램을 만들 수 있는 환경을 말합니다.
* IDLE 사용 단점 : 컴퓨터에 깔아야됨 (내가 다 해놓음 ^^;)
* IDLE 사용 장점 : 파일 저장이 쉽고, Colab에서 안되는 것들이 됨, (ColabTurtle 말고 그냥 Turtle로 색깔 있는 그림 그릴 수 있음)

```python
from tkinter import *

# 모양생성

window = Tk() #최상위 윈도우

l1 = Label(window, text="화씨")
l2 = Label(window, text="섭씨")
#l3 = 

l1.pack()
l2.pack()
#

e1 = Entry(window)
e2 = Entry(window)
#e3 = 
e1.pack()
e2.pack()
#


b1 = Button(window,text="화씨==>섭씨")
b2 = Button(window,text="섭씨==>화씨")
#b3 = 

b1.pack()
b2.pack()

# 클릭, 입력 등의 작업 처리
window.mainloop()
```
