### 첫번째 import, from부분은 필요한 도구를 불러오는 부분입니다.

*   text에 빈도수를 고려하여 아무 단어나 입력해보고 (구분은 띄어쓰기)
*   실행버튼을 눌러 결과를 확인해보며
*   각 줄이 의미하는 바를 추리해봅시다.
*   폰트 에러 메시지가 뜬다면? 저번시간에 수행했던것처럼 런타임 -> 런타임 다시 시작 -> 예 클릭 후 차분하게 위부터 다시 실행해봅시다.



```
from wordcloud import WordCloud, STOPWORDS

text = "수능 수능 집 사대부고 강원도 코로나 점심 컴퓨터실 하하하하 호호호호" #시각화 대상이 되는 단어들을 띄어쓰기로 구별하여 입력 
print(text) #위에서 입력한 것을 출력해줌 
wordcloud = WordCloud(font_path='/usr/share/fonts/truetype/nanum/NanumGothicEco.ttf', background_color='white').generate(text) #worldcloud라는 변수(공간)에 WordCloud라는 도구를 사용하여 시각화 설정

plt.figure(figsize=(15,15)) #이미지 사이즈 지정 
plt.imshow(wordcloud, interpolation='lanczos') #wordcloud안에 들어있는 설정값을 이용하여 시각화를 하고 그 결과값을 저장, 뒤에 있는 것은 부드러움의 정도 (신경X)
plt.axis('off') #x y 축 숫자 제거
plt.savefig('/content/gdrive/MyDrive/PL(2021)/pro3/worldcloud1.png') #위에서 만든 결과값을 파일로 만듦 
plt.show()  #최종 결과를 그림으로 보여줌
```

### 워드클라우드 배경 마스크 설정하기

* 지금까지 우리가 만든 워드 클라우드는 재미없는 사각형 모양을 가지고 있었다.
* 워드클라우드의 마스크 기능을 사용하면 우리가 원하는 모양으로 단어 구름을 만들 수 있다.
* 만들고 싶은 단어구름의 기반이 되는 1) 텍스트 2) 배경 마스크 그림파일 3) https://www.remove.bg/ 사이트를 이용한 편집을 이용하면
* 내가 만들고 싶은 모양의 단어 구름들을 만들 수 있다.

```
import numpy as np
from PIL import Image

#마스크 설정을 위해 필요한 도구를 불러옴

from wordcloud import WordCloud, STOPWORDS #시각화를 위한 도구인 wordcloud(전체)에서 시각화를 담당하는 WordCloud도구와, 특정 단어를 제거할 수 있는 STOPWORDS 도구를 불러옴

f = open('/content/gdrive/MyDrive/PL(2021)/pro3/my.txt','r') # (메모장을 열어놓어 자신이 관심있는 텍스트(뉴스기사, 자기소개서, 발표 대본 등)을 저장하고 경로를 정확하게 입력
path = '/content/gdrive/MyDrive/PL(2021)/pro3/sixd-removebg-preview.png' #마스크 대상이 되는 파일의 경로

word_mask = np.array(Image.open(path))

data = f.read()
#print(data)

spwords=set(STOPWORDS) # 기본적으로 많이 쓰는 단어 (제외할 단어) # ex)'should', "wouldn't", "how's", "i'm", 'then', etc ..
spwords.add('') #제외하고 싶은 단어 추가 * (2) 시각화에서 큰 의미가 없는 단어들을 최소 3개 이상 채우기
spwords.add('') #제외하고 싶은 단어 추가
spwords.add('') #제외하고 싶은 단어 추가


wordcloud = WordCloud(max_font_size=200, font_path='/usr/share/fonts/truetype/nanum/NanumGothicEco.ttf',    #폰트 크기, 폰트 경로, 제외할 단어, 배경색, 크기.generate(대상이 되는 글자)
                     stopwords=spwords,
                     background_color='gray',
                     width=1200,height=800,
                    mask = word_mask,
                    contour_width=0.5,
                    contour_color='black').generate(data)

plt.figure(figsize=(10,8)) #크기 지정 
plt.imshow(wordcloud) #이미지 생성 
plt.tight_layout(pad=0) #여백 설정
plt.axis('off') #x,y축 값 삭제 
plt.savefig('/content/gdrive/MyDrive/PL(2021)/pro3/maskword.png') #DPL에 mywordmap.png로 파일 생성 
plt.show() #생성된 시각화 보여주기 
f.close() #파일 사용 마지막
```

### 나만의 워드 클라우드 마스크 설정하기

```
import numpy as np
from PIL import Image

from wordcloud import WordCloud, STOPWORDS #시각화를 위한 도구인 wordcloud(전체)에서 시각화를 담당하는 WordCloud도구와, 특정 단어를 제거할 수 있는 STOPWORDS 도구를 불러옴

f = open('','r') #  메모장을 열어놓어 자신이 관심있는 텍스트를 저장하고 경로 설정 
path = '' #배경이 될 마스크의 경로 설정

word_mask = np.array(Image.open(path))

data = f.read()
#print(data)

spwords=set(STOPWORDS) # 기본적으로 많이 쓰는 단어 (제외할 단어) # ex)'should', "wouldn't", "how's", "i'm", 'then', etc ..
spwords.add('') #제외하고 싶은 단어 추가 * (2) 시각화에서 큰 의미가 없는 단어들을 최소 3개 이상 채우기
spwords.add('') #제외하고 싶은 단어 추가
spwords.add('') #제외하고 싶은 단어 추가


wordcloud = WordCloud(max_font_size=500, font_path='/usr/share/fonts/truetype/nanum/NanumGothicEco.ttf',    #폰트 크기, 폰트 경로, 제외할 단어, 배경색, 크기.generate(대상이 되는 글자)
                     stopwords=spwords,
                     background_color='white',
                     width=1920,height=1080,
                    mask = word_mask,
                    contour_width=0.5,
                    contour_color='',
                    colormap='').generate(data)

plt.figure(figsize=(10,8)) #크기 지정 
plt.imshow(wordcloud) #이미지 생성 
plt.tight_layout(pad=0) #여백 설정
plt.axis('off') #x,y축 값 삭제 
plt.savefig('') 결과물이 생성될 경로 결정
plt.show() #생성된 시각화 보여주기 
f.close() #파일 사용 마지막
```
